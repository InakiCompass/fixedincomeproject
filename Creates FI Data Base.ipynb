Imports Fundamental Data from Excel<a id='i_f'></a>
[<a href='#back'>back</a>]

```python
import pandas as pd
pd.set_option('display.max_columns', 500)
#pd.set_option('display.max_rows', 500)
f_data = pd.read_excel('CREDIT LATAM.xlsm',sheet_name='SUMMARY',skiprows=2)
f_data = f_data.iloc[:,1:25]
f_data = f_data.drop(['Mkt. Cap.', 'Fitch', 'S&P', 'Moodys', '1Q19', 'Ticker CIQ'], axis=1)
f_data = pd.melt(f_data, id_vars=['Company', 'Country', 'Sector', 'Character', 'Period', 'Source', 'Bloomberg ticker'], var_name="Account", value_name="Value")
f_data = f_data.sort_values(by='Company')
f_data
```

Imports ZSpread Data from Excel<a id='i_z'></a>
[<a href='#back'>back</a>]

```python
zs_data = pd.read_excel('RF_python.xlsm',sheet_name='ZS datos',skiprows=25,header=[0,1,2])
t = zs_data.T
t.columns = t.iloc[0]
#t = t.drop(['Security Name'])
#pd.melt(t, id_vars =[('Security Name' 'Issuer Name','Value')], var_name='Fecha', value_name='valor')
t = zs_data.T
t.reset_index(inplace=True)
t.columns = t.iloc[0]
del t['Value']
zs_data = pd.melt(t, id_vars =['Security Name','Issuer Name'], var_name= 'Fecha', value_name='Value')
zs_data = zs_data.drop(0)
zs_data = zs_data.rename(columns={'Issuer Name': 'Ticker'})
zs_data
```

Imports Market Data from Excel<a id='i_m'></a>

```python
mkt_data = pd.read_excel('RF_python.xlsm',sheet_name='BD')
mkt_data = mkt_data.drop(['@', 'Unnamed: 87', 'Unnamed: 89', 'Portfolio', 'Obs'], axis=1)
mkt_data = pd.melt(mkt_data, id_vars=['Ticker', 'Security Name', 'Cntry Of Risk', 'Crncy', 'Callable', 'Floater', 'Fixed', 'Sinkable', 'Industry Sector', 'Industry Group', 'Rtg Sp', 'Rtg Moody', 'Rtg Fitch', 'Rtg Bb Comp', 'S&P Transformado', 'Moodys Transformado', 'Fitch Transformado', 'Rtg CG', 'RTG GRUPO', 'Ticker.1', 'Listed Equity?', 'Listed Equity 2', 'Pais+LE', 'RTG+LE', 'ID_ISIN', 'Ttrace', 'Trace+Isin', 'Ticker 2'], var_name="Account", value_name="Value")
mkt_data = mkt_data.sort_values(by='Ticker')
mkt_data
```
