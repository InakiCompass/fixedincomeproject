```python
import pandas as pd
pd.set_option('display.max_columns', 500)
#pd.set_option('display.max_rows', 500)
f_data = pd.read_excel('CREDIT LATAM.xlsm',sheet_name='SUMMARY',skiprows=2)
f_data = f_data.iloc[:,1:25]
f_data = f_data.drop(['Mkt. Cap.', 'Fitch', 'S&P', 'Moodys', '1Q19', 'Ticker CIQ'], axis=1)
f_data = pd.melt(f_data, id_vars=['Company', 'Country', 'Sector', 'Character', 'Period', 'Source', 'Bloomberg ticker'], var_name="Account", value_name="Value")
f_data = f_data.sort_values(by='Company')
f_data
```
