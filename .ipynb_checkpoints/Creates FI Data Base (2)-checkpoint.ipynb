```python
import sqlite3
from sqlite3 import Error
import pandas as pd
import urllib
import sqlalchemy
```

# Fundamental

```python
pd.set_option('display.max_columns', 500)
#pd.set_option('display.max_rows', 500)
f_data = pd.read_excel('CREDIT LATAM.xlsm',sheet_name='SUMMARY',skiprows=2)
f_data = f_data.iloc[:,1:25]
f_data = f_data.drop(['Mkt. Cap.', 'Fitch', 'S&P', 'Moodys', '1Q19', 'Ticker CIQ', 'Bloomberg ticker' ], axis=1)
f_data = pd.melt(f_data, id_vars=['Company', 'Sector', 'Character', 'Period', 'Source'], var_name="Account", value_name="Value")
f_data = f_data.rename(columns={'Company': 'f_data_issuer_Company', 'Sector': 'f_data_Sector', 'Character': 'f_data_Character', 'Period': 'f_data_Period', 'Source': 'f_data_Source', 'Bloomberg ticker': 'f_data_Bloomberg_ticker', 'Account': 'f_data_Account', 'Value': 'f_data_Value' })
f_data = f_data.sort_values(by='f_data_issuer_Company')
f_data = f_data.dropna(subset=['f_data_Value'])
f_data = f_data.drop_duplicates()

##################
#DATAFRAME TO SQL#
##################

conn = sqlite3.connect('Fixed_Income_DB500.db')
f_data.to_sql(name='f_data', con=conn, if_exists = 'append', index=False)
conn.commit()

##################
#      QUERY     #
##################

dff = pd.read_sql_query("SELECT * FROM f_data", conn)
conn.close()
dff = dff[["f_data_Account","f_data_Value"]]
dff
f_data
```

# ZSpread

```python
zs_data = pd.read_excel('RF_python.xlsm',sheet_name='ZS datos',skiprows=25,header=[0,1,2])
t = zs_data.T
t.columns = t.iloc[0]
#t = t.drop(['Security Name'])
#pd.melt(t, id_vars =[('Security Name' 'Issuer Name','Value')], var_name='Fecha', value_name='valor')
t = zs_data.T
t.reset_index(inplace=True)
t.columns = t.iloc[0]
del t['Value']
zs_data = pd.melt(t, id_vars =['Security Name','Issuer Name'], var_name= 'Fecha', value_name='Value')
zs_data = zs_data.drop(0)
zs_data = zs_data.rename(columns={'Issuer Name': 'z_spread_Ticker', 'Security Name': 'z_spread_sec_Security_Name', 'Fecha': 'z_spread_Date' , 'Value': 'z_spread_Value'})
zs_data = zs_data.dropna()
zs_data = zs_data[['z_spread_Ticker','z_spread_sec_Security_Name', 'z_spread_Date', 'z_spread_Value' ]]
zs_data

####################
# DATAFRAME TO SQL #
####################

zs_data.dtypes
zs_data['z_spread_Ticker']= zs_data['z_spread_Ticker'].astype('str')
zs_data['z_spread_sec_Security_Name']= zs_data['z_spread_sec_Security_Name'].astype('str')
zs_data['z_spread_Date']= zs_data['z_spread_Date'].astype('str')
zs_data['z_spread_Value']= zs_data['z_spread_Value'].astype('str')

conn = sqlite3.connect('Fixed_Income_DB500.db')
zs_data.to_sql(name='z_spread', con=conn, if_exists = 'append', index=False)
conn.commit()
 

##################
#      QUERY     #
##################

dfz = pd.read_sql_query("SELECT * FROM z_spread", conn)
conn.close()
#dfz
zs_data
```
