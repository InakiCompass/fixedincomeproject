```python
import sqlite3
import pandas as pd
import urllib
import sqlalchemy
conn=sqlite3.connect(r'Z:\RESEARCH\Empresas\Bases de Datos\Data_Chile.db')
```

```python
c=conn.cursor()

#create issuers table
c.execute('''CREATE TABLE IF NOT EXISTS issuers (
    Issuer_Compass_Id TEXT PRIMARY KEY,
    Issuer_Compass_Alias TEXT NOT NULL,
    Issuer_Local_Industry TEXT NOT NULL,
    Issuer_Country TEXT NOT NULL,
    Issuer_Funct_Curncy TEXT NOT NULL)''')

#create accounts table
c.execute('''CREATE TABLE IF NOT EXISTS accounts (
    Account_Id INTEGER PRIMARY KEY,
    Acc_DVAI_Id TEXT,
    Acc_DVAF_Id TEXT,
    Acc_Nombre_Cuenta TEXT NOT NULL,
    Account_Name TEXT NOT NULL,
    Account_Type TEXT NOT NULL)''')

#create estimates table
c.execute('''CREATE TABLE IF NOT EXISTS estimates (
    Est_Id INTEGER PRIMARY KEY AUTOINCREMENT,
    Est_Issuer_Compass_Id TEXT,
    Est_Account_Id TEXT NOT NULL,
    Est_Period TEXT,
    Est_Forecast_Date NUMERIC,
    Est_Value NUMERIC,
    FOREIGN KEY (Est_Issuer_Compass_Id) REFERENCES issuers (Issuer_Compass_Id),
    FOREIGN KEY (Est_Account_Id) REFERENCES accounts (Account_Id))''')

#create hist table
c.execute('''CREATE TABLE IF NOT EXISTS hist_data (
    Hist_Id INTEGER PRIMARY KEY AUTOINCREMENT,
    Hist_Issuer_Compass_Id TEXT,
    Hist_Account_Id TEXT NOT NULL,
    Hist_Period TEXT NOT NULL,
    Hist_Value NUMERIC,
    FOREIGN KEY (Hist_Issuer_Compass_Id) REFERENCES issuers (Issuer_Compass_Id),
    FOREIGN KEY (Hist_Account_Id) REFERENCES accounts (Account_Id))''')

#create securities table
c.execute('''CREATE TABLE IF NOT EXISTS securities (
    Secur_Id TEXT PRIMARY KEY,
    Sec_Issuer_Compass_Id TEXT NOT NULL,
    Sec_Bbg_Ticker TEXT,
    Sec_Isin TEXT,
    Sec_Trade_Curncy TEXT NOT NULL,
    FOREIGN KEY (Sec_Issuer_Compass_Id) REFERENCES issuers (Issuer_Compass_Id))''')

#create positions table
c.execute('''CREATE TABLE IF NOT EXISTS positions (
    Pos_Id IDENTITY PRIMARY KEY,
    Pos_Sec_Id TEXT,
    Pos_Instrument TEXT NOT NULL,
    Pos_Date TEXT,
    Pos_Value INT,
    FOREIGN KEY (Pos_Sec_Id) REFERENCES securities (Security_Id))''')

#create quarters table
c.execute('''CREATE TABLE IF NOT EXISTS quarters (
    Q_Id INTEGER PRIMARY KEY AUTOINCREMENT,
    Q_Issuer_Compass_Id TEXT NOT NULL, 
    Q_Data_Type TEXT NOT NULL,
    Q_Institution TEXT,
    Q_Period_Type TEXT NOT NULL,
    Q_Forecast_Date NUMERIC,
    Q_Year INT NOT NULL,
    Q_Period INT NOT NULL,
    Q_Account_Id INT NOT NULL,
    Q_Value NUMERIC,
    FOREIGN KEY (Q_Account_Id) REFERENCES accounts (Account_Id),
    FOREIGN KEY (Q_Issuer_Compass_Id) REFERENCES issuers (Issuer_Compass_Id))''')

#create banks table
c.execute('''CREATE TABLE IF NOT EXISTS banks_results (
    B_Id INTEGER PRIMARY KEY AUTOINCREMENT,
    B_Issuer_Compass_Id TEXT NOT NULL,
    B_Year INT NOT NULL,
    B_Month INT NOT NULL,
    B_SBIF_Account_Id INT NOT NULL,
    B_Account_Id INT,
    b_Value NUMERIC,
    FOREIGN KEY (B_Account_Id) REFERENCES accounts (Account_Id),
    FOREIGN KEY (B_Issuer_Compass_Id) REFERENCES issuers (Issuer_Compass_Id))''')
```
